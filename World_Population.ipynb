```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('World_Population.xlsm',sheet_name=2)
df.head(10)
```

```python
df.info()
```

```python
for column in df.columns:
    print(f'====={column}=====')
    display(df[column].value_counts())
```

```python
df[['Total Population']].hist()
```

```python
df.groupby(['Continent'])['Total Population'].max().to_frame()
```

```python
df.groupby(['Continent'],sort=False)['Total Population'].sum().to_frame()
```

```python
df.groupby(['Continent'],sort=False)['Total Population'].count().to_frame()
```

```python
df.groupby(['Continent'],sort=False).agg({'Total Population':['count','sum']})
```

```python
df.groupby(['Total Population'],sort=True)['Country/Territory'].sum().to_frame().head(3)
```

```python
df.sort_values('Total Population',inplace=True)
x=df['Country/Territory']
y=df['Total Population']
plt.figure(figsize=(70,50))
plt.barh(x,y,color='yellowgreen')

plt.title('Total Populasi di Seluruh Dunia',size=16)
plt.xlabel('Jumlah Populasi',size=14)
plt.ylabel('Negara Bagian',size=14)
plt.xticks(size=12)
plt.yticks(size=12)
```

```python
df.sort_values('Total Population',inplace=True)
x=df['Continent']
y=df['Total Population']
plt.figure(figsize=(20,5))
plt.barh(x,y,color='yellowgreen')

plt.title('Total Populasi di Seluruh Dunia',size=16)
plt.xlabel('Jumlah Populasi',size=14)
plt.ylabel('Continent',size=14)
plt.xticks(size=12)
plt.yticks(size=12)
```
